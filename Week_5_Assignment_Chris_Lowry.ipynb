```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# View the first few rows of the dataset
titanic.head()
```

From the dataset present on the titanic, the first thing I noticed of the survivors, is all of them are female. Of the members of the third class, the only survivor was the woman. This could indicate women and children were first to go onto life boats before adult males. to find further info on if the class led to any saftey we would need males from first class and not just third class.

```python
# Create a histogram for the fare distribution
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.hist(titanic['fare'],bins=60, color='red',edgecolor='white')
plt.title('Fare Distribution of Titanic Passengers')
plt.xlabel('Fare(dollars)')
plt.ylabel('count')
plt.grid(axis='x')
plt.xticks([0, 50, 100, 150, 200, 250, 300, 350, 400])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic['fare'], bw_adjust=0.5, fill=True, color="blue", alpha=0.5)
plt.title('Kernel Density Estimate of Fares for Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.xlim(0, 600)
plt.show()
```

The outliers I notice in fares are there just a few passengers with fares above 500, while most are in between 0 and 100. After looking at another chart or the table we can see if the amount paid had anything to do with survival rating and if paying more gave first priority to lifeboats potentially.

```python
plt.figure(figsize=(10, 6))

# KDE for survived
sns.kdeplot(titanic[titanic['survived'] == 1]['fare'], bw_adjust=0.7, fill=True, color="blue", alpha=0.7, label='Survived')

# KDE for non-survived
sns.kdeplot(titanic[titanic['survived'] == 0]['fare'], bw_adjust=0.7, fill=True, color="pink", alpha=0.7, label='Non-Survived')

# Set plot title and labels
plt.title('Kernel Density Estimate of Fares for Survivors and Non-survivors Titanic Passengers')
plt.xlabel('Fares')
plt.ylabel('Density')

# Add a legend to distinguish the lines
plt.legend()

# Set x-axis limits for clarity
plt.xlim(0, 600)

# Show the plot
plt.show()
```

The fares did seem to have an impact on survival suggesting those of higher economic status were saved first or fsater because they paid more while those who had lower fares were non-survivors.

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Plot the class 1 histogram, mirrored to the left (negative weights)
plt.hist(titanic[titanic['pclass'] == 1]['fare'], bins=30, label='pclass 1', alpha=0.5, color='blue', orientation='vertical')

# Plot the class 2 histogram on the right side (default)
plt.hist(titanic[titanic['pclass'] == 2]['fare'], bins=30, label='pclass 2', alpha=0.5, color='red', orientation='vertical')

# Plot the class 3 histogram on the left side (default)
plt.hist(titanic[titanic['pclass'] == 3]['fare'], bins=30, label='pclass 3', alpha=0.5, color='yellow', orientation='vertical')

# Add labels and title
plt.xlabel('Fares')
plt.ylabel('Pclass')
plt.title('Fares compared to class of Passenger')

# Customize ticks on the x-axis to show positive values for both sides
plt.yticks([0, 1, 2, 3])

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Plot the class 1 histogram, mirrored to the left (negative weights)
plt.hist(titanic[titanic['pclass'] == 1]['survived'], bins=30, label='pclass 1', alpha=0.5, color='blue', orientation='vertical')

# Plot the class 2 histogram on the right side (default)
plt.hist(titanic[titanic['pclass'] == 2]['survived'], bins=30, label='pclass 2', alpha=0.5, color='red', orientation='vertical')

# Plot the class 3 histogram on the left side (default)
plt.hist(titanic[titanic['pclass'] == 3]['survived'], bins=30, label='pclass 3', alpha=0.5, color='yellow', orientation='vertical')

# Add labels and title
plt.xlabel('Survival')
plt.ylabel('Pclass')
plt.title('Survival rate compared to class of Passenger')

# Customize ticks on the x-axis to show positive values for both sides
plt.yticks([0, 1, 2, 3])

# Add a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

The people who are in class 1 paid a higher fare, and just like fare to survival rate distribution, those in class 1 had higher survival and then class2 had the sceond most, and class 3 had the least survivers. this shows the statement made earlier the class and fare had big implications to the survivor rate.

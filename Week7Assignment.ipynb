```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())
```

```python
iris.info()
```

```python
iris.describe()
```

###Pairplot

```python
# Pairplot will create all scatter plots for the numerical columns
sns.pairplot(iris,hue="species",diag_kind="kde")
plt.title("Pairplot of Iris Dataset")
plt.show
```

####Bar Plot

```python
# Bar plot of species vs average sepal length
#sns.barplot(x="species", y="sepal_length", data=iris)

plt.bar(iris['species'], iris['sepal_length'])
# Add title and labels
plt.title('Average Sepal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Sepal length (cm)')

# Show the plot
plt.show()
```

####Box Plot

```python
# Box plot of species vs distribution of petal widths
#sns.boxplot(x="species", y="petal_widths", data=iris)

plt.figure(figsize=(10, 6))
sns.boxplot(x=iris['species'], y=iris['petal_width'])

# Add title and labels
plt.title('Petal Widths for Each Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')

# Show the plot
plt.show()
```

####Combined Bar Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Bar plot for sepal length by species
sns.barplot(x='species', y='sepal_length', data=iris, ax=axes[0])
plt.xlabel('Species')
plt.ylabel('Sepal length (cm)')


# Bar plot for petal length by species
sns.barplot(x='species', y='petal_length', data=iris, ax=axes[1])
plt.title('Average Petal and Sepal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Petal length (cm)')


# Show the plot
plt.show()
```

####Violin Plot

```python
# Create the violin plot
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.xlabel('Species')
plt.ylabel('Sepal length (cm)')
plt.title('Distribution of Sepal Length by Species')

plt.show()
```

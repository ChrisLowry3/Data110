```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")  # Sets the style for the plots
```

```python
# Data from the table
crimes = ['Aggrevated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
amounts = [264.1, 250.7, 1347.2, 318.7, 5.7, 38, 66.5]

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(crimes, amounts, color='red')

# Add titles and labels
plt.title('Offenses per 100,000 people 2023')
plt.xlabel('Crimes')
plt.ylabel('Amount per 100,000 People')

# Add only horizontal grid lines with a dashed style
plt.grid(axis='y', linestyle='-', color='gray')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plot a horizontal bar chart
plt.figure(figsize=(8, 5))
colors='red', 'green', 'blue', 'purple', 'orange', 'yellow', 'black'
plt.barh(crimes, amounts, color= colors, zorder=3)  # Horizontal bars with zorder=1

# Data from the table
crimes = ['Aggrevated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
amounts = [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]

# Add titles and labels
plt.title('Changes in Offenses per 100,000 people 2022-2023')
plt.xlabel('Changes in Amount per 100,000 People')
plt.ylabel('Crimes')


# Get the current axis using gca()
ax = plt.gca()

# Set x-axis limits to ensure -16 is included (since it's a horizontal bar chart, the x-axis shows the amounts)
ax.set_xlim(-14, 14)

# Set x-axis ticks to appear at -14, -12, -8, -4, 0, 4, 8, 12, 14
plt.xticks([-14, -12, -8, -4, 0, 4, 8, 12, 14])

# Add only horizontal grid lines with a dashed style
plt.grid(axis='y', linestyle='-', color='gray')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create the DataFrame
data = {'Crimes': ['Aggrevated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
        'Amounts': [-3.3, -8.1, -4.9, 12, -12, -9.8, -0.8]}
dm = pd.DataFrame(data)

# Plot the horizontal bar chart
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 6))
colors = ['green', 'green', 'green', 'red', 'green', 'green', 'green']
plt.barh(dm['Crimes'], dm['Amounts'], color=colors)

# Add titles and labels
plt.title('Change in Crime Amounts per 100,000 people 2022-2023')
plt.xlabel('Increase or Decrease in amount of each crime 2022-2023')
plt.ylabel('Crime')

# Get the current axis using gca()
ax = plt.gca()

# Set x-axis limits to ensure -16 is included (since it's a horizontal bar chart, the x-axis shows the amounts)
ax.set_xlim(-14, 14)

# Set x-axis ticks to appear at -14, -12, -8, -4, 0, 4, 8, 12, 14
plt.xticks([-14, -12, -8, -4, 0, 4, 8, 12, 14])

# Add only horizontal grid lines with a dashed style
plt.grid(axis='y', linestyle='-', color='gray')

# Show the plot
plt.tight_layout()
plt.show()
```

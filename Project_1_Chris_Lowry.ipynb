```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import kagglehub

# Download latest version
path = kagglehub.dataset_download("heesoo37/120-years-of-olympic-history-athletes-and-results")

print("Path to dataset files:", path)
```

```python
df=pd.read_csv(path+"/athlete_events.csv")
df.head()
```

```python
# Create a medal count per Sex
medal_count = df.groupby('Sex')['Medal'].count().reset_index()

# Create bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better visibility
sns.barplot(x=medal_count['Sex'], y=medal_count['Medal']) # Pass data as named arguments

# Add title and labels
plt.title('Medal Count per Sex')
plt.xlabel('Sex')
plt.ylabel('Medal Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
sex_counts = df['Sex'].value_counts()

# Create bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better visibility
# Pass sex_counts Series as data and specify x and y
sns.barplot(x=sex_counts.index, y=sex_counts.values)

# Add title and labels
plt.title('Number of Sex in Olympics')
plt.xlabel('Sex')
plt.ylabel('Sex Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
#Showing the exact number of each sex in Olympics
total_sex_counts  = df['Sex'].value_counts()
print(total_sex_counts)
```

```python
total_medal_counts  = df.groupby('Sex')['Medal'].value_counts()
print(total_medal_counts)
```

```python
#Create grouped bar plot showing medal count by sex and in which seasons
medal_count = df.groupby(['Sex', 'Season'])['Medal'].count().reset_index()

sns.barplot(x='Sex', y='Medal', hue='Season', data=medal_count)

# Add title and labels
plt.title('Medal Count by Sex and Season')
plt.xlabel('Sex')
plt.ylabel('Medal Count')
```

```python
#Create grouped bar plot showing medal count by sex and in which seasons
medal_count = df.groupby(['Year', 'Sex'])['Medal'].count().reset_index()

sns.barplot(x='Year', y='Medal', hue='Sex', data=medal_count)

# Add title and labels
plt.title('Medal Count by Sex and Year')
plt.xlabel('Year')
plt.ylabel('Medal Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#Create grouped bar plot showing medal count by sex and in which seasons
medal_count = df.groupby(['Year', 'Season'])['Medal'].count().reset_index()

sns.barplot(x='Year', y='Medal', hue='Season', data=medal_count)

# Add title and labels
plt.title('Medal Count by Season and Year')
plt.xlabel('Year')
plt.ylabel('Medal Count')
plt.xticks(rotation=90)
plt.show()
```
